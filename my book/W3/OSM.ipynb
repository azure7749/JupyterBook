# 台灣鐵道車站分佈圖

## 作業內容
利用overpass api取得全島營運中鐵路車站位置，並加以分類以呈現出捷運, 台鐵及高鐵車站的位置

```python
import folium
import requests
```

```python
def fetch_overpass_data(query):
    overpass_url = "http://overpass-api.de/api/interpreter"
    response = requests.get(overpass_url, params={'data': query})
    data = response.json()
    return data
```

```python
metro_query = """
[out:json];
area["name:en"="Taiwan"]->.searchArea;
(
  node["station"="subway"]["subway"="yes"](area.searchArea);
);
out center;
"""
tra_query = """
[out:json];
area["name:en"="Taiwan"]->.searchArea;
(
  node["railway"="station"]["network:en"="TRA"](area.searchArea);
);
out center;
"""
hsr_query = """
[out:json];
area["name:en"="Taiwan"]->.searchArea;
(
node["railway"="station"]["network:en"="Taiwan High Speed Rail"](area.searchArea);
);
out center;
"""
metro_data = fetch_overpass_data(metro_query)
tra_data = fetch_overpass_data(tra_query)
hsr_data = fetch_overpass_data(hsr_query)
```

```python
taiwan_center = [23.6978, 120.9605]
m = folium.Map(location=taiwan_center, zoom_start=8)

def add_markers(data, map_obj, color, station_type):
    for element in data['elements']:
        lat = element['lat']
        lon = element['lon']
        name = element['tags'].get('name', 'Unnamed Station')
        folium.Marker(
            location=[lat, lon],
            popup=f"{station_type}: {name}",
            icon=folium.Icon(color=color)
        ).add_to(map_obj)

add_markers(metro_data, m, 'blue', 'Metro Station')
add_markers(tra_data, m, 'green', 'Train Station')
add_markers(hsr_data, m, 'red', 'HSR Station')


m    
```
